3. What is the earliest year of birth?

```python
earliest_year_birth = df_name["birthYear"].min()
print(f"Earliest year of birth: {earliest_year_birth}")
```

4. How many years ago was this person born?

```python
time_actuel = time.time()
year_actuel = datetime.fromtimestamp(time_actuel).year
print(f"This person was born {year_actuel - int(earliest_year_birth)} years ago")
```

5. Using only the data in the data set, determine if this date of birth correct. 


```python
oldest_person = df_name[df_name["birthYear"] == earliest_year_birth]
# .iloc[0] to avoid a warning
Death_Year = int(oldest_person["deathYear"].iloc[0])
age_person = Death_Year - int(earliest_year_birth)
print(f"Age of the person at death: {age_person} years old (Death Year: {int(oldest_person["deathYear"].iloc[0])})")
```

7. What is the most recent data of birth?  

```python
# 'coerce' replace any elements that are not numbers (especially in our case the '\N') by NaN
numeric_birth_year = pd.to_numeric(df_name["birthYear"], errors='coerce')
```

```python
# max() ignores  NaN but not \N
recent_year_birth = int(numeric_birth_year.max())
print(f"Most recent year of birth: {recent_year_birth}")
```

8. What percentage of the people do not have a listed date of birth?


```python
no_birthYear = numeric_birth_year.isna().sum()
total = df_name["nconst"].nunique()
```

```python
percentage_missing = (no_birthYear/total)*100
print(f"{percentage_missing:.2f}% of people have no birth date.")
```

11. List of all of the genres represented.

12. What is the higest rated comedy "movie" in the dataset? Note, if there is a tie, the tie shall be broken by the movie with the most votes .


```python
df_movie = df_title_basics[df_title_basics["titleType"] == "movie"]
```

```python
df_movie_comedy = df_movie[df_movie["genres"].str.contains("Comedy", na=False)]
```

```python
# List of tconst of comedy movies
df_comedy_movie_ratings = df_title_ratings[df_title_ratings["tconst"].isin(df_movie_comedy["tconst"])]
```

```python
df_comedy_movie_ratings_sorted = df_comedy_movie_ratings.sort_values(["averageRating", "numVotes"], ascending=False)
```

```python
df_comedy_movie_ratings_sorted.set_index("tconst").head()
```

```python
tconst_best_rated_movie_comedy = df_comedy_movie_ratings_sorted.iloc[0]["tconst"]
```

```python
movie = df_title_basics[df_title_basics["tconst"] == tconst_best_rated_movie_comedy]
```

```python
movie.set_index("tconst").head()
```

```python
print(f"The best rated and most voted comedy movie is {movie["primaryTitle"].values[0]}")
```
